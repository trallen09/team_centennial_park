```python
df1=pd.read_csv('Resources/City_Zhvi_1bedroom.csv')
df2=pd.read_csv('Resources/City_Zhvi_2bedroom.csv')
df3=pd.read_csv('Resources/City_Zhvi_3bedroom.csv')
df4=pd.read_csv('Resources/City_Zhvi_4bedroom.csv')
df5=pd.read_csv('Resources/City_Zhvi_5BedroomOrMore.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5], ignore_index=True)
```

```python
df.set_index(['Bedrooms','RegionID','SizeRank','RegionName','RegionType','StateName','State','Metro','CountyName'], inplace=True)
```

```python
df = df.stack().reset_index()
```
